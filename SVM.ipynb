```python
import pandas as pd
import numpy as np
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.model_selection import train_test_split, cross_val_score
```

# Q1. Forest_Fire

Question:- Prepare a classification model using SVM for salary data.

```python
data = pd.read_csv('forestfires.csv')
data
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
X = data.iloc[:,2:30]
X
```

```python
data['size_category'].value_counts()
```

```python
data['size_category']= data['size_category'].map({'small':0,'large':1})
data
```

```python
X = data.iloc[:,2:30]
Y = data.iloc[:,30]
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size = 0.20)
```

```python
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
clf = SVC(kernel ='linear',C= 0.1, gamma = 50)
clf.fit(x_train , y_train)
y_pred = clf.predict(x_test)
acc = accuracy_score(y_test, y_pred) * 100
print("Accuracy =", acc)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
clr = classification_report(y_test,y_pred)
print(clr)
```

There is no need for grid search as we have got the maximum accuracy.

# Q2. Salary_Data

```python
Question:- Classify the Size_Categorie using SVM.
```

```python
d= pd.read_csv('SalaryData_Train.csv')
d
```

```python
d1 = pd.read_csv('SalaryData_Test.csv')
d1
```

```python
print('Columns of a salary_train data are : {}'.format(d.columns))
print('Columns of a salary_train data are : {}'.format(d1.columns))
```

```python
string_col=['workclass','education','maritalstatus','occupation','relationship','race','sex','native']
```

```python
from sklearn import preprocessing
label_encoder=preprocessing.LabelEncoder()
for i in string_col:
    d[i]=label_encoder.fit_transform(d[i])
    d1[i]=label_encoder.fit_transform(d1[i])
```

```python
train_x=d.iloc[0:500,0:13]
train_y=d.iloc[0:500,13]
test_x=d1.iloc[0:300,0:13]
test_y=d1.iloc[0:300,13]
```

```python
model_linear=SVC(kernel='linear')
model_linear.fit(train_x,train_y)
```

```python
train_pred_lin=model_linear.predict(train_x)
```

```python
test_pred_lin=model_linear.predict(test_x)
```

```python
train_lin_acc=np.mean(train_pred_lin==train_y)
test_lin_acc=np.mean(test_pred_lin==test_y)
print('train accuracy is',(train_lin_acc))
print('test accuracy is',(test_lin_acc))
```

```python
confusion_matrix(test_y,test_pred_lin)
```

```python
clr = classification_report(test_y,test_pred_lin)
print(clr) 
```
